```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
vanilla = pd.read_csv('./results/cifar10/res-56-1-t-bn/sgd/step/ort_both_none_0.01_0.01_0.01_0.01_app_seed1_False_log.csv')
snr = pd.read_csv('./results/cifar10/res-56-1-t-bn/sgd/step/ort_both_sin_srip_0.1_0.1_0.1_0.1_app_seed1_False_log.csv')
srip = pd.read_csv('./results/cifar10/res-56-1-t-bn/sgd/step/ort_both_srip_0.1_0.1_0.1_0.1_app_seed1_False_log.csv')
srip2 = pd.read_csv('./results/cifar10/res-56-1-t-bn/sgd/step/ort_both_srip_0.1_0.1_0.1_0.1_app_seed5_False_log.csv')
norm = pd.read_csv('./results/cifar10/res-56-1-t-bn/sgd/step/ort_both_norm_0.01_0.01_0.01_0.01_app_seed1_False_log.csv')
iso = pd.read_csv('./results/cifar10/res-56-1-t-bn/sgd/step/ort_both_isometry_0.1_0.1_0.1_0.1_app_seed4_False_log.csv')
```

```python
vanilla['train_accuracy'][:10].values
```

```python
plt.rcParams.update({'font.size': 15})

plt.plot(vanilla['train_accuracy'][:10].values, '.-', color='blue')
plt.plot(snr['train_accuracy'][:10].values,'.-' ,color = 'black')
plt.plot(srip['train_accuracy'][:10].values,'.-', color = 'purple')
plt.plot(srip2['train_accuracy'][:10].values,'.-', color = 'red')
plt.plot(norm['train_accuracy'][:10].values,'.-', color = 'green')
plt.plot(iso['train_accuracy'][:10].values,'.-', color = 'orange')

plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.savefig('train.pdf')
```

```python
plt.rcParams.update({'font.size': 15})

plt.plot(vanilla['test_accuracy'][:10].values, '.-', color='blue')
plt.plot(snr['test_accuracy'][:10].values,'.-' ,color = 'black')
plt.plot(srip['test_accuracy'][:10].values,'.-', color = 'purple')
plt.plot(srip2['test_accuracy'][:10].values,'.-', color = 'red')
plt.plot(norm['test_accuracy'][:10].values,'.-', color = 'green')
plt.plot(iso['test_accuracy'][:10].values,'.-', color = 'orange')

plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.savefig('test.pdf')
```
